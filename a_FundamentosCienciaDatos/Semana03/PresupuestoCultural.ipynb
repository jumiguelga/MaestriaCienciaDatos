# EDA _ Eventos Culturales Medellín

## Paso 1: Identificación del actual Data Set:

```python
import pandas as pd

eventos_df = pd.read_csv('./datos_eventos_01.csv')
eventos_df.info()
eventos_df.head()
```

## Paso 2: Declaración correcta de los tipos de datos.
Debido a que se evidenció que algunos datos, como el estrato, no fueron reconocidos de forma correcta, hemos 
efectuado algunos ajustes en cuanto a los tipos de datos.

```python
eventos_df_copy = eventos_df.copy()

eventos_df_copy['Lugar'] = eventos_df_copy['Lugar'].astype('category')
eventos_df_copy['Estrato_Zona'] = eventos_df_copy['Estrato_Zona'].astype('category')
eventos_df_copy['Categoria'] = eventos_df_copy['Categoria'].astype('category')
eventos_df_copy['Fecha'] = pd.to_datetime(eventos_df_copy['Fecha'])
eventos_df_copy.info()
```

```python
# Returns True if there are any duplicates in 'column_name'
has_duplicates = eventos_df_copy['ID_Evento'].duplicated().any()
print(has_duplicates)
```
